```python
import nltk
nltk.download("punkt_tab")
```

```python
file = open("/sentence.txt","r",encoding="utf-8")
text = file.read()
print(text)
```

```python
# 1. Given some text file, tokenize them into sentences using NLTK

from nltk.tokenize import sent_tokenize
sentence_tokens = sent_tokenize(text)
print(sentence_tokens)
```

```python
# 2. Given some text file, tokenize them into words using NLTK
from nltk.tokenize import word_tokenize
word_tokens = word_tokenize(text)
print(word_tokens)
```

```python
# 3. Remove stop words ( is, a, the, it....)
# Stop words are the one which are mostly used and are not useful for understanding the language using embeddings. If stop words are not removed then the computation will be waste because we will be creating a table with each word as a column, these stopwords are useless here and if not removed increases the compuation power

from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words("english"))
filtered = [word for word in word_tokens if word.lower() not in stop_words]
print(filtered)
```

```python
# 4. After removing stop words also words like ", . / [ ] " etc ( We want only alphabets) count the frequency of each word

filtered_words = []
for word in word_tokens:
  if word.isalpha() and word.lower() not in stop_words:
    filtered_words.append(word)
print(filtered_words)

# computing frequencies of each word
from nltk.probability import FreqDist
freq_map = FreqDist(filtered_words)
for word, freq in freq_map.items():
  print(word,":",freq)
```

```python
# 5. Remove Rare words ( using word frequency ) - word frequency less than a particular threshold is removed
threshold = 2
fully_filtered=[]
for word, freq in freq_map.items():
  if freq > threshold:
    fully_filtered.append(word)
print(fully_filtered)
```
